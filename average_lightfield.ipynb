```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_lightfield(filepath):
    
    lf = np.load(filepath)
    lf = np.nan_to_num(lf)
    
    print("Loaded lightfield with dtype:", lf.dtype, "and shape:", lf.shape)
    
    return lf

lightfield = load_lightfield("original_lightfield.npy")
```

```python
def get_average_lightfield(lf):
    u_num, v_num, *_ = lf.shape
    
    lf = lf*256 if lf.dtype == np.float32 else lf
    lf = lf.astype(np.uint16)
    
    sum_lf = np.sum(lf, axis=0)
    sum_lf = np.sum(sum_lf, axis=0)
    
    avg_lf = sum_lf/(u_num*v_num)
    avg_lf = avg_lf.astype(np.uint8)
    # print("Average lightfield shape:", avg_lf.shape)
    
    return avg_lf
    
average_lf = get_average_lightfield(lightfield)
```

```python
import tqdm
for i in tqdm.tqdm(range(10)):
    average_lf = get_average_lightfield(lightfield)
```

```python
plt.figure(figsize= (10,10))
plt.imshow(average_lf)
```

```python
def display_average_lightfield(path):
    lf = load_lightfield(path)
    average_lf = get_average_lightfield(lf)
    plt.figure(figsize= (10,10))
    plt.imshow(average_lf)    
```

```python
display_average_lightfield("original_lightfield.npy")
```
