```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

n_images = 6
width= 880
height = 660
```

```python
result = np.empty((n_images, height,width,3), dtype=np.uint8)

for i, image_path in enumerate(sorted([path for path in (Path(".") / "results_rendering").iterdir()
                                if str(path).endswith(".bmp")])):
    print(str(image_path))
    image_data = Image.open(image_path)
    result[i] = np.array(image_data)
```

```python
fig, ax = plt.subplots(2, 3)
fig.set_size_inches(15, 8)

fig.suptitle("Views of train")

for i in range(n_images):
    curr_ax = ax[i//3][i%3]
    im = curr_ax.imshow(result[i], aspect= "auto")
    curr_ax.set_title(str(40+2*i))
    curr_ax.axis("off")

plt.savefig("./figs/plot_6.png")
```
